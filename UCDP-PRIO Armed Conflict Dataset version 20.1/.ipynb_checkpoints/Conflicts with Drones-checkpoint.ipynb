```python
# NOT USED IN PAPER


import numpy as np
import pandas as pd

all_events = pd.read_csv('../UCDP Georeferenced Event Dataset (GED) Global version 20.1/ged201.csv') 
all_conflicts = pd.read_csv('ucdp-prio-acd-201.csv') 
```

```python
all_events
```

```python
# add drone column if match
all_events.loc[all_events['source_headline'].str.contains('drone', case = 0) == 1, 'has_drone'] = '1'
all_events.loc[all_events['source_headline'].str.contains('drone', case = 0) == 0, 'has_drone'] = '0'

all_events.loc[all_events['source_article'].str.contains('drone', case = 0) == 1, 'has_drone'] = '1'
all_events.loc[all_events['source_article'].str.contains('drone', case = 0) == 0, 'has_drone'] = '0'

all_drone_events = all_events.loc[all_events['has_drone'] == "1"]
all_drone_events
```

```python
# collect conflict id for each drone event

count = 0
conflict_ids_from_drone_evens = {}

for index, row in all_drone_events.iterrows():
    count = count+1
    #print("****", row['has_drone'], row['source_headline'], "-----", row['source_article'], " Best Death:", row["best"])
    #print(row['conflict_new_id'])
    conflict_ids_from_drone_evens[row['conflict_new_id']] = 1

    #if (count == 500):
     #   break

# all the keys are conflict ids with drone evens
conflict_ids_from_drone_evens
```

```python
# get those conflicts
#all_conflicts

conflicts_with_drones = pd.DataFrame()


for key, value in conflict_ids_from_drone_evens.items():
    print("----- EINTRÄGE FÜR KONFLIKT ID: "+str(key))
    #print(all_conflicts[all_conflicts['conflict_id'] == key & all_conflicts['year'] == 2001])
    #print(all_conflicts.loc[all_conflicts['conflict_id'] == key and all_conflicts['conflict_id'] != 0])
    
    one = all_conflicts.loc[(all_conflicts['conflict_id'] == key) & (all_conflicts['year'] >= 2001)]
    #print(one.conflict_id)
        
    conflicts_with_drones = one.append(conflicts_with_drones)
    #conflicts_with_drones.append(one)

(conflicts_with_drones)
```
