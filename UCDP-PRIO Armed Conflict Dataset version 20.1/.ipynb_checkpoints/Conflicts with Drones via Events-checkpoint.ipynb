```python
'''
Collect all Events with "drone" in text
Get IDs of Conflicts
Mark conflicts as "drone conflict"
'''

import numpy as np
import pandas as pd

# load all events & all conflicts
all_events = pd.read_csv('../../UCDP Georeferenced Event Dataset (GED) Global version 20.1/ged201.csv')
all_conflicts = pd.read_csv('../ucdp-prio-acd-201.csv')
```

```python
# dict with word to search. value will be amount of hits (value not used here)
# "unmanned" was a test: had no results. there were no other synonyms for drone
search_list = {
  "drone": 0,
  "unmanned": 0
}

# if column exist already drop it, just in case
if 'has_drone' in all_events.columns:
    all_events.drop(columns=["has_drone"])

# set all to 0 as default
all_events.loc[all_events['source_headline'] != 1, 'has_drone'] = '0'

# search
for key, value in search_list.items():
    print("------ Search for: "+key)

    # set to 1 if string found
    all_events.loc[all_events['source_headline'].str.contains(key, case = 0) == 1, 'has_drone'] = '1'
    all_events.loc[all_events['source_article'].str.contains(key, case = 0) == 1, 'has_drone'] = '1'

# after search: create single df only with drone events
all_drone_events = all_events.loc[all_events['has_drone'] == "1"]
#print(all_drone_events)
```

```python
conflict_ids_from_drone_events_in_a_year = {}

# Counted prepareation for drone usage in a given year
# split per year -> so: only count it as drone conflict for a year, if there was an event in a year
for year in range(2001, 2020):
    #print(year)

    conflict_ids_from_drone_events_in_a_year[year] = {}

    for index, row in all_drone_events.loc[all_drone_events.year == year].iterrows():
    #print("****", row['has_drone'], row['source_headline'], "-----", row['source_article'], " Best Death:", row["best"])
    #print(row['conflict_new_id'])
        conflict_ids_from_drone_events_in_a_year[year][row['conflict_new_id']] = 1

    #conflict_ids_from_drone_events_in_a_year[year] = 1

conflict_ids_from_drone_events_in_a_year
```

```python
# Mark Conflict a drone conflict (in a year: means, the conflict is only a drone conflict if drone fighting was there)

# default all false
all_conflicts.loc[all_conflicts.conflict_id != 0, "is_drone_conflict_in_this_year"] = False

# key is year, value is dict. with conflict IDs
for key, value in conflict_ids_from_drone_events_in_a_year.items():
    print("----- EINTRÄGE FÜR JAHR: "+str(key))

    # key2 is conflict ID, value2 is just 1
    for key2, value2 in value.items():
        print("--------- KONFLIKT ID: "+str(key2))

        # set is_drone_conflict_overall true for all conflicts with certain Year AND ID
        all_conflicts.loc[(all_conflicts.year == key) & (all_conflicts.conflict_id == key2), "is_drone_conflict_in_this_year"] = True
```

```python
# Graph: Drone Conflicts in a year
# Chapter 4.2 Fig. 5
# Code: ufzzrrg44

# total amount of conflicts in a given year, used for the %
totals_per_year = all_conflicts.loc[all_conflicts.year >= 2004].value_counts(all_conflicts['year'])

ax = all_conflicts.loc[all_conflicts.year >= 2004].groupby('is_drone_conflict_in_this_year').year.value_counts().unstack(0).plot(kind='bar', stacked=True, figsize=(13,10))


ax.legend(["Anteil Konflikte ohne Einsatz von Drohnen", "Anteil Konflikte mit Einsatz von Drohnen"])
ax.set_ylabel('Anzahl der Konflikte')
ax.set_xlabel('Jahr')

#df2.plot(figsize=(20,10))

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()

    if (height != 0):
        ax.text(x+width/2,
                y+height/2,
                #'{:.0f} %'.format(totals_per_year[x+0.25]),
                #'{:.0f} %'.format(height),
                str(round(height/totals_per_year[x+0.25]*100))+" %",
                rotation=90,
                horizontalalignment='center',
                verticalalignment='center')

```

```python
# list with all drone conflicts
# Used for Table 1 in Chapter 4.2

list_with_drone_conflicts = all_conflicts.loc[all_conflicts.year >= 2001].loc[all_conflicts.is_drone_conflict_in_this_year == True]
list_with_drone_conflicts
list_with_drone_conflicts.to_csv("list_with_drone_conflicts.csv")


for year in range(2001, 2020):
    print("------ Drone Conflicts in Year "+str(year))

    for index, row in all_conflicts.loc[all_conflicts.year == year].loc[all_conflicts.is_drone_conflict_in_this_year == True].iterrows():
        print(row['conflict_id'])
        #print(row['type_of_conflict'])
```
