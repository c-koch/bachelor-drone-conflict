```python
import numpy as np
import pandas as pd
```

```python
dfAllData = pd.read_csv('../ged201.csv') 
```

```python
dfAllData

count = 0

for index, row in dfAllData.iterrows():
    count = count+1
    print("****", row['has_drone'], row['source_headline'], "-----", row['source_article'], " Best Death:", row["best"])
    if (count == 50):
        break
```

```python
# Mark row if drone-event
dfAllData.loc[dfAllData['source_headline'].str.contains('drone', case = 0) == 1, 'has_drone'] = '1'
dfAllData.loc[dfAllData['source_headline'].str.contains('drone', case = 0) == 0, 'has_drone'] = '0'

dfAllData.loc[dfAllData['source_article'].str.contains('drone', case = 0) == 1, 'has_drone'] = '1'
dfAllData.loc[dfAllData['source_article'].str.contains('drone', case = 0) == 0, 'has_drone'] = '0'

dfAllDrones = dfAllData.loc[dfAllData['has_drone'] == "1"]
#dfAllDrones
```

```python
# Bar chart with only with drone-events (Chapter 4.1 Fig. 2)
# Final Paper uses stata code
dfAllDrones = dfAllDrones.sort_values(by=['year'], ascending=True)
dfAllDrones['year'].value_counts(sort = False).plot(kind='bar')
```

```python
# Not used, killed by drones per year
s2 = dfAllDrones.groupby(['year']).sum()
s2['best'].plot(kind='bar')
```
